```python
import scanpy as sc

adata1 = sc.read_h5ad("data/raw/dataset1.h5ad")
adata2 = sc.read_h5ad("data/raw/dataset2.h5ad")

print("adata1:", adata1)
print("adata2:", adata2)
print("adata1 shape:", adata1.shape)
print("adata2 shape:", adata2.shape)
```

```python
import scanpy as sc

print("adata1:", adata1)
print("adata2:", adata2)

print("\nadata1 cell_type:")
print(adata1.obs["cell_type"].value_counts())

print("\nadata2 cell_type:")
print(adata2.obs["cell_type"].value_counts())

print("\nadata2 disease:")
print(adata2.obs["disease"].value_counts())

if "disease_state" in adata2.obs.columns:
    print("\nadata2 disease_state:")
    print(adata2.obs["disease_state"].value_counts())

# subset normal cells from adata2
adata2_normal = adata2[adata2.obs["disease"].astype(str) == "normal"].copy()
print("\nadata2_normal:", adata2_normal)

# shared cell types
ct1 = set(adata1.obs["cell_type"].astype(str).unique())
ct2 = set(adata2_normal.obs["cell_type"].astype(str).unique())

print("\nadata1 cell types:", sorted(ct1))
print("\nadata2_normal cell types:", sorted(ct2))
print("\nshared cell types:", sorted(ct1.intersection(ct2)))

# gene names
print("\nadata1 var_names preview:")
print(adata1.var_names[:10])

print("\nadata2 var_names preview:")
print(adata2.var_names[:10])

print("\nadata1 feature_name preview:")
print(adata1.var["feature_name"].head(10))

print("\nadata2 feature_name preview:")
print(adata2.var["feature_name"].head(10))

# shared genes
common_genes = adata1.var_names.intersection(adata2_normal.var_names)
print("\nnumber of shared genes:", len(common_genes))

# layers/raw
print("\nadata1 layers:", list(adata1.layers.keys()))
print("adata2 layers:", list(adata2.layers.keys()))
print("adata1 raw exists:", adata1.raw is not None)
print("adata2 raw exists:", adata2.raw is not None)
```
